## UPDATE (15FEB2024):

However, now I'm stuck trying to figure out how to get the average temperature added into these data frames since I'm missing 545,686 values within these rows which is a significant amount of data when trying to find patterns. My first thought would be to see if I have enough of the Max/Min temperatures, I could calculate the average based off of that information and then input it there. Lets take a look at how that would be accomplished!

````python
import numpy as np

# Calculate the average temperature based on min and max temperature
merged_df['avg_temp_imputed'] = merged_df[['min_temp_c', 'max_temp_c']].mean(axis=1)

# Replace missing values in 'avg_temp_c' column with the calculated average
merged_df['avg_temp_c'].fillna(merged_df['avg_temp_imputed'], inplace=True)

# Drop the temporary column used for imputation
merged_df.drop(columns=['avg_temp_imputed'], inplace=True)
````

## Explanation of Above code:

Let's break down the code step by step:

```python
import numpy as np
```
This line imports the NumPy library and assigns it the alias 'np'. NumPy is a popular Python library used for numerical computing. We'll use NumPy later to perform mathematical operations.

```python
merged_df['avg_temp_imputed'] = merged_df[['min_temp_c', 'max_temp_c']].mean(axis=1)
```
Here, we're creating a new column in our DataFrame called 'avg_temp_imputed'. We're calculating the average temperature for each row by taking the mean (average) of the 'min_temp_c' and 'max_temp_c' columns. The `.mean(axis=1)` part specifies that we want to calculate the mean across columns (axis=1) for each row.

```python
merged_df['avg_temp_c'].fillna(merged_df['avg_temp_imputed'], inplace=True)
```
This line fills in missing values in the 'avg_temp_c' column with the calculated averages from the 'avg_temp_imputed' column. The `fillna()` method replaces NaN (missing) values with the specified value. In this case, we're using the values from the 'avg_temp_imputed' column to fill in the missing values in the 'avg_temp_c' column. The `inplace=True` parameter modifies the DataFrame in place, meaning the changes are made directly to the DataFrame `merged_df`.

```python
merged_df.drop(columns=['avg_temp_imputed'], inplace=True)
```
Finally, we drop the temporary column 'avg_temp_imputed' that we created for imputation purposes. We no longer need this column once we've filled in the missing values in the 'avg_temp_c' column.

Overall, this code takes advantage of the available temperature data ('min_temp_c' and 'max_temp_c') to estimate missing average temperature values. It's a practical approach to handling missing data in this context.
